````python
!docanalysis --project_name fin_climate_justice --make_section --search_section INT, RES, CON, DIS --dictionary COUNTRY --output fin_climatejustice.csv
````

## **Displaying wordcloud for country list made from csv (output) in Colab environment**

````python
# Import necessary libraries
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Import the files module if using Google Colab
try:
    from google.colab import files
except ImportError:
    files = None  # files will remain None if not in a Colab environment

# Load your CSV file into a DataFrame
csv_file_path = '/content/fin_climate_justice/fin_climatejustice.csv'  # Replace with your actual file path
df = pd.read_csv(csv_file_path)

# Extract text from "column 0"
text_data = ' '.join(df['0'].astype(str).tolist())

# Split the text into words and count their frequencies
word_list = text_data.split()
word_freq = Counter(word_list)  # Count the frequency of each word

# Generate a word cloud, where the size depends on word frequency
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate_from_frequencies(word_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Save the word cloud as an image file
wordcloud_file_path = '/content/fin_list_country.png'
wordcloud.to_file(wordcloud_file_path)

# Download the word cloud image if in Colab
if files:
    files.download(wordcloud_file_path)
````
